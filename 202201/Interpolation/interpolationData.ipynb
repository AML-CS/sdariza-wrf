```python
import iris
import numpy as np
```

```python
nc_data = iris.load('../mergeNETCDF/all.nc')
```

```python
sample_points = [('longitude', np.linspace(0, 359, 360)),
                 ('latitude',  np.linspace(-90, 90, 181))]
```

```python
nc_data
```

```python
for indx, _ in enumerate(nc_data):
    nc_data[indx] = nc_data[indx].interpolate(sample_points, iris.analysis.Linear())
```

```python
nc_data
```

```python
iris.save(nc_data, 'allDataInterpolated.nc')
```
