```python
import pygrib
import os
```

```python
def getvbreq(): # get variables by levels needed
    level = [1000, 925, 850,  700, 600, 500, 400, 300, 250, 200, 150, 100, 70, 50, 30, 20, 10, 1, 0]
    for grib2File in os.listdir('../../workspace/grib-data/'):
        grbs = pygrib.open('../../workspace/grib-data/'+grib2File)
        grb = grbs.read()
        grb = grbs.select(shortName=['t','u','v','w', 'gh', 'r', 'sp', 'st'], typeOfLevel=['isobaricInhPa','surface', 'depthBelowLandLayer'], level = level)
#         grb = [ g for i, g in enumerate(grb) if i != 109 and i > 5]
        grb = grb[6:]
        messages = []
        for g in grb:
            messages.append(g.tostring())
        grbs.close()
        grbout = open('../../workspace/gribs-data/'+grib2File,'wb')

        for msg in messages:
            grbout.write(msg)
        grbout.close()
        print(f'Variables taken from {grib2File} successfully')
    print('Process done!')
```

```python
getvbreq()
```
