```python
allNC = nc.Dataset('allData.nc','w', format="NETCDF4")
```

```python
allNC.createDimension('lon',len(airNC.variables['lon']))
allNC.createDimension('lat',len(airNC.variables['lat']))
allNC.createDimension('time',None)
allNC.createDimension('level',len(airNC.variables['level']))
```

```python
allNC
```

```python
time = allNC.createVariable('time', 'f8','time')
lon = allNC.createVariable('lon', 'f4','lon')
lat = allNC.createVariable('lat', 'f4','lat')
level = allNC.createVariable('level','f4', 'level')
air = allNC.createVariable('air', 'f4', ('time', 'level','lat', 'lon'))
uwnd = allNC.createVariable('uwnd', 'f4', ('time', 'level','lat', 'lon'))
vwnd = allNC.createVariable('vwnd', 'f4', ('time', 'level','lat', 'lon'))
```

```python
allNC
```

```python
print(airNC)
print(uwndNC)
print(vwndNC)
```

```python
time[:] = airNC.variables['time'][:]
lon[:] = airNC.variables['lon'][:]
lat[:] = airNC.variables['lat'][:]
level[:] = airNC.variables['level'][:]
air[:] =airNC.variables['air'][:]
uwnd[:] =uwndNC.variables['uwnd'][:]
vwnd[:] =vwndNC.variables['vwnd'][:]
```

```python
allNC.description = "merge datasets - airTemp, uwnd, vwnd"
allNC.history = 'merged: '+dt.datetime.today().strftime("%d/%m/%y")
```

```python
allNC.variables['time'].__dict__
```

```python
time.units = airNC.variables['time'].units
lon.units = airNC.variables['lon'].units
lat.units = airNC.variables['lat'].units
level.units = airNC.variables['level'].units
air.units =airNC.variables['air'].units
uwnd.units =uwndNC.variables['uwnd'].units
vwnd.units =vwndNC.variables['vwnd'].units
```

```python
time.long_name = "Time"
time.units = 'hours since 1800-1-1 00:00:0.0'
```
